### Relations between independent and dependent variables.

##### 1) First finding relation between continuous Numerical variables with respect to profit.

```python
sns.regplot(x='Postal Code',y='Profit', data = df)
plt.title("Profit V/s  Postal code")
```

```python
df[['Postal Code','Profit']].corr()
```

#### Observation of graph and correlation value :- 
It indicate that postal code has negative linear correlation with profit.  


```python
sns.regplot(x='Sales',y='Profit', data = df)
plt.title("Sales V/s  Profit")
```

```python
df[['Sales','Profit']].corr()
```

#### Observation of graph and co-relation value :-
It indicate that Sales has positive linear correlation with profit.  

```python
sns.regplot(x='Quantity',y='Profit', data = df)
plt.title("Quantity V/s  Profit")
```

```python
df[['Quantity','Profit']].corr()
```

#### Observation of graph and correlation value :- 
It indicate that Quantity has positive linear correlation with profit.  

```python
sns.regplot(x='Discount',y='Profit', data = df)
plt.title("Discount V/s  Profit")
```

```python
df[['Discount','Profit']].corr()
```

#### Observation of graph and correlation value :-
It indicate that Discount has negative linear correlation with profit.  

```python
sns.heatmap(df.corr(method='pearson'),annot=True)
```

Above heatmap clearly visualize co-relations between all variables. and we clearly visualize that Sales,Quantity are positively related to profit if sales get increases profit also increases and discount is negatively related to profit.
